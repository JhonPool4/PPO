```python
from myenvs.inverted_pendulum.trajectory_control import InvertedPendulum
from ppo import PPO
from torch.optim import Adam
import torch.nn as nn
import torch
import gym
```

```python
# create model
env=InvertedPendulum()
#env = gym.make('Pendulum-v0')

# with double tanh
# with 
hyperparameters = {'epochs': 1000,
                   'max_ep_len':400,
                   'steps_per_epoch':4000,                    
                   'coef_ent':0.001,
                   'target_kl':0.01,                   
                   'pi_lr':3e-4,
                   'vf_lr':1e-3,
                   'train_pi_iters':50,
                   'train_vf_iters':50,
                   'save_freq':100,
                   'flag_render':'False',
                   'create_new_training_data':'False'}
ppo=PPO(env, **hyperparameters)
```

```python
# load model: 3000

ppo.ac_model.load_state_dict(torch.load('./training/ppo_ac_model.pth'))
```

```python
# train agent
env.flag_render=False
ppo.learn()
```

```python
# test model

episodes = 5
env.flag_render= True
env.FPS = 100

for episode in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0 
    env.iter=episode
    steps=0
    while not done:
        env.render()
        action = ppo.ac_model.act(torch.from_numpy(obs)) #env.action_space.sample()
        obs, reward, done, info = env.step(action)
        score+=reward
        steps+=1
        
        #print(f'action: {action}')
    print(f'Episode:{episode}, Score:{score/steps}')
env.close()  
```

```python
import numpy as np
def combined_shape(length, shape=None):
    if shape is None:
        return (length,)
    return (length, shape) if np.isscalar(shape) else (length, *shape)   
```

```python
import numpy as np
import scipy.signal
import torch

def discount_cumsum(x, discount):
    """
    magic from rllab for computing discounted cumulative sums of vectors.
    input: 
        vector x, 
        [x0, 
         x1, 
         x2]
    output:
        [x0 + discount * x1 + discount^2 * x2,  
         x1 + discount * x2,
         x2]
    """
    return scipy.signal.lfilter([1], [1, float(-discount)], x[::-1], axis=0)[::-1]
```

```python
import torch
import numpy as np
path_slice=slice(2,5)

device='cuda:0'

rew_buf = torch.ones(10, dtype=torch.float32).to('cuda:0')
val_buf = torch.ones(10, dtype=torch.float32).to('cuda:0')
ret_buf = torch.ones(10, dtype=torch.float32).to('cuda:0')
adv_buf = torch.ones(10, dtype=torch.float32).to('cuda:0')
for i in range(10):
    rew_buf[i]=i
    val_buf[i]=i

last_value = 0

print(f"rew_buf: {rew_buf}, {rew_buf.shape} \n")

rews = np.append(rew_buf[path_slice].cpu().numpy(), last_value)
vals = np.append(val_buf[path_slice].cpu().numpy(), last_value)
print(f"rews: {rews}, {type(rews)} \n")
print(f"vals: {vals}, {type(vals)} \n")
```

```python
deltas = rews[:-1] + 1* vals[1:] - vals[:-1]
adv_buf[path_slice] = torch.as_tensor(discount_cumsum(deltas, 1).copy(), dtype=torch.float32).to(device)

# the next line computes rewards-to-go, to be targets for the value function
temp = discount_cumsum(rews, 1)[:-1]
ret_buf[path_slice] = torch.as_tensor(temp.copy(), dtype=torch.float32).to(device) 

print(f"adv: {adv_buf} \n")
print(f"ret: {ret_buf} \n")
```
